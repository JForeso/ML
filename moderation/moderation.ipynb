```python
from urllib.request import urlretrieve
import pandas as pd

# Download the parquet table
table_url = f'https://huggingface.co/datasets/poloclub/diffusiondb/resolve/main/metadata.parquet'
urlretrieve(table_url, 'metadata.parquet')

# Read the table using Pandas
df = pd.read_parquet('metadata.parquet')
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
df['prompt_nsfw'].max()
```

```python
df1 = df[:100000]
df2 = df[100001:200001]
df1.loc[df['prompt_nsfw'] >= 0.4, "prompt_nsfw"] = "Harmful"
df1.loc[df['prompt_nsfw']  < 0.4, "prompt_nsfw"] = "Harmless"
df2.loc[df['prompt_nsfw'] >= 0.4, "prompt_nsfw"] = "Harmful"
df2.loc[df['prompt_nsfw']  < 0.4, "prompt_nsfw"] = "Harmless"
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df1['prompt'])
```

```python
y = df1['prompt_nsfw']
```

```python
neigh = KNeighborsClassifier(n_neighbors=31)
```

```python
neigh.fit(X, y)
```

```python
X2 = vectorizer.fit_transform(df2['prompt'])
y_pred = neigh.predict(X)
```

```python
accuracy_score(y, y_pred)
```

```python
precision_score(y, y_pred, average='macro')
```

```python
recall_score(y, y_pred, average='macro')
```

```python
f1_score(y, y_pred, average='macro')
```
